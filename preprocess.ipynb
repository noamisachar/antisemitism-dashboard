```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/dima_dataset.csv', sep=';')
```

```python
df1 = df1[df1['year'] == 2014]
df1 = df1[['country', 'region', 'adl_index']]
df1 = df1.dropna(subset=['adl_index'])
country_mapping = {
    'Bosnia_Herzegovina': 'Bosnia and Herzegovina',
    'Costa_Rica': 'Costa Rica',
    'Cote_d_Ivoire': "Cote d'Ivoire",
    'CzechRepublic': 'Czech Republic',
    'Dominican_Rep': 'Dominican Republic',
    'Egypt': 'Egypt, Arab Rep.',
    'Iran': 'Iran, Islamic Rep.',
    'Laos': 'Lao PDR',
    'New_Zealand': 'New Zealand',
    'Russia': 'Russian Federation',
    'Saudi_Arabia': 'Saudi Arabia',
    'South_Africa': 'South Africa',
    'South_Korea': 'Korea, Rep.',
    'Trinidad_and_Tobago': 'Trinidad and Tobago',
    'UAE': 'United Arab Emirates',
    'UK': 'United Kingdom',
    'US': 'United States',
    'Venezuela': 'Venezuela, RB',
    'Yemen': 'Yemen, Rep.'
}
df1['country'] = df1['country'].map(country_mapping).fillna(df1['country'])
```

```python
df1
```

```python
df2 = pd.read_csv('data/EdStatsData.csv')
```

```python
years = ['2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
df2 = df2[df2['Country Name'].isin(df1['country'])]
df2 = df2.dropna(subset=years, how='all')
df2 = df2[df2.groupby('Indicator Name')['Indicator Name'].transform('count') >= len(df2['Country Name'].unique())]
df2['value'] = df2[years].mean(axis=1, skipna=True)
df2 = df2[['Country Name', 'Indicator Name', 'value']]
df2 = df2.pivot(index='Country Name', columns='Indicator Name', values='value').reset_index()
df2 = df2.rename(columns={'Country Name': 'country'})
```

```python
df2
```

```python
df3 = pd.read_excel('data/democracy_indices.xlsx', sheet_name='interpolate-missing-year')
```

```python
df3.columns = df3.iloc[0]
df3 = df3.iloc[1:]
country_mapping_dem = {
    'Egypt': 'Egypt, Arab Rep.',
    'Iran': 'Iran, Islamic Rep.',
    'Lao': 'Lao PDR',
    'Russia': 'Russian Federation',
    'South Korea': 'Korea, Rep.',
    'Venezuela': 'Venezuela, RB',
    'Yemen': 'Yemen, Rep.'
}

df3['name'] = df3['name'].map(country_mapping_dem).fillna(df3['name'])
df3 = df3[df3['name'].isin(df1['country'])]
df3 = df3[df3['year'] == 2014]
df3 = df3.drop(['year', 'Change in democracy index'], axis=1)
df3 = df3.rename(columns={'name': 'country'})
df3[df3.columns[2:]] = df3[df3.columns[2:]].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
df3
```

```python
df = pd.merge(pd.merge(df1, df2, on='country'), df3, on='country')
```
